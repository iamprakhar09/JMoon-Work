```python
import pandas as pd
from pandas import DataFrame, Series
import numpy as np

df = pd.read_csv("kickstarter_2017_Feb_subset.csv")
df.head()
```

```python
# Find blurb and name length including and excluding stopwords
from nltk.corpus import stopwords
cachedStopWords = stopwords.words("english")

def text_length(row):
    try:
        name_str = row['name']
        name_len = len(name_str.split())
        name_clean = ' '.join(word for word in name_str.split() if word not in cachedStopWords)
        name_len_clean = len(name_clean.split())

        blurb_str = row['blurb']
        blurb_len = len(blurb_str.split())
        blurb_clean = ' '.join(word for word in blurb_str.split() if word not in cachedStopWords)
        blurb_len_clean = len(blurb_clean.split())

        return name_len, name_len_clean, blurb_len, blurb_len_clean
    except:
        error = "NaN"
        return error, error, error, error
```

```python
# create empty columns
df['name_len'] = np.nan
df['name_len_clean'] = np.nan
df['blurb_len'] = np.nan
df['blurb_len_clean'] = np.nan

for i in df.index.values.tolist():
    returned = text_length(df.iloc[i])
    df = df.set_value(i, 'name_len', returned[0])
    df = df.set_value(i, 'name_len_clean', returned[1])
    df = df.set_value(i, 'blurb_len', returned[2])
    df = df.set_value(i, 'blurb_len_clean', returned[3])

df.head()
```

```python
df.columns
```

```python
# convert all times to datetime
import datetime

def get_time(value):
    return datetime.datetime.fromtimestamp(value)
    # return datetime.datetime.fromtimestamp(value).strftime('%Y-%m-%d %H:%M:%S')

date_col_list = ['deadline','state_changed_at','created_at','launched_at']
for c in date_col_list:
    df[c] = df[c].apply(get_time)
df.head()
```

```python
# Create new columns for weekday
def get_weekday(value):
    return value.strftime('%A')
    
for c in date_col_list:
    c_name = c + "_weekday"
    df[c_name] = df[c].apply(get_weekday)

df['deadline_weekday'].head()
```

```python
# Create new column for month, day and year
def get_month(value):
    return value.month

def get_day(value):
    return value.day

def get_yr(value):
    return value.year

def get_hr(value):
    return value.hour

for c in date_col_list:
    c_month = c + "_month"
    df[c_month] = df[c].apply(get_month)
    c_day = c + "_day"
    df[c_day] = df[c].apply(get_day)
    c_yr = c + "_yr"
    df[c_yr] = df[c].apply(get_yr)
    c_hr = c + "_hr"
    df[c_hr] = df[c].apply(get_hr)

df.iloc[0]
```

```python
# create new column for number of days between specific campaign milestones
df['create_to_launch'] = df['launched_at'] - df['created_at']
df['launch_to_deadline'] = df['deadline'] - df['launched_at']

# if campaign is successful, number of days between launch and statechange
def find_success_len(row):
    if row['state'] == "successful":
        try:
            return row['state_changed_at'] - row['launched_at']
        except:
            return np.nan
    else:
        return np.nan
df['launch_to_state_change'] = df.apply(find_success_len, axis = 1) 
```
